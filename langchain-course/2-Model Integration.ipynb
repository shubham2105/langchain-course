### Streaming and Batching

### Streaming

Most models can stream their output while its being generated. By displaying output progressively, streaming significantly improves user experience, particularly for longer responses.Calling stream() returns an iterator that yeilds output chunks as they are produced.
You can use a loop to process each chunk in real time.

```python
model.invoke("write a 200 words paragraph on AI")
```

```python
for chunk in model.stream("Why do parrots have colourfurl feathers?"):
    print(chunk.text,end="|", flush=True)
```

### Batch

Batching is a collection of independent requests to a model which can significantly imporve performance and reduce costs, as the processing is done parallely.

```python
questions = model.batch ([
    "Why do parrots have colourful feathers?",
    "How do aeroplanes fly?",
    "What is quantum computing?",
    "Who is Messi",
    "What is football",
    "What is Computer Science"
], config={"max_concurrency": 2})  

for questions in questions:
    print(questions)
```
