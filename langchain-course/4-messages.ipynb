### Messages

Messages are the fundanetal unit of context for models in Langchain. They represent the input and output of models, carrying both the
content and metadata needed to represent the state of a conversation when interacting with LLM. Messages are object that contain:

1. Role: Identifies the message type (e.g. system, user)
2. Content: Represens the actual content of the message (like text, images, audio, documents, etc)
3. Metadata: Optional fields such as response information, messages IDs and token usages

Langchain provides a standard message type that works all model providers, ensuring consistent behaviour regardless of the model being called.

```python
import os 
from langchain.chat_models import init_chat_model

os.environ["MISTRAL_API_KEY"] = os.getenv("MISTRAL_API_KEY")
print(os.getenv("MISTRAL_API_KEY"))

model = init_chat_model("mistral-medium-3-5", model_provider="mistralai")
model.invoke("Who is Leo Messi?")
```

### Text Prompts

Text prompts are string - ideal for a straightforward generation tasks where you dont need to retain conversation history.

```python
model.invoke("What is Machine Learning?")
```

### Use text prompts when: 
    1. You have a single, standalone request
    2. You dont need conversatio history
    3. you want minimal code complexity

### Message Prompts

Alternatively, you can pass a list of messages to the model by providing a list of message objects.

Message types:

    1. System Messages - Tells the model how to behave and provide comtext for interactions.
    2. Human Messages - Represents  user input and interactions with model
    3. Ai Message - Responses generated by the model, including text content, tool calls and metadata
    4. Tool message - Represents the outputs of tool calls

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage("You are an angry person"),
    HumanMessage("What is Reinforcment Learning ")
]
response = model.invoke(messages)
response.content
```
